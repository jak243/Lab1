```python
import numpy as np
alpha = 60
erd = 0.04
def rad(x):
    return x*np.pi/180.0
def error(erd,d,alpha):
    errorr = np.sqrt((erd**2)*(0.5*np.cos((rad(d+alpha))/2.0)/np.sin(rad(alpha)/2.0))**2)
    return errorr
#errorr = np.sqrt(erd**2(0.5*np.cos((rad(d+alpha))/2.0)/np.sin(rad(alpha)/2.0)))
def n(d,alpha):
    n = np.sin(rad(d+alpha)/2)/np.sin(rad(alpha)/2)
    return n
```

```python
dmin = np.array([52.15,50.32,49.47,49.22,48.47,48.24,48.03,47.62])
print(dmin)
error = error(erd,dmin,alpha)
print(error)
n = n(dmin,alpha)
print(n)
```

```python
wvlngth = np.array([404.7,435.9,486.1,546.1,577.0,599.0,589.3,656.2])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.optimize as op
```

```python
plt.errorbar(wvlngth,n,yerr=error, fmt='o')
plt.title('Graph 1 Wavelength and n')
plt.xlabel("Wavelength in nm")
plt.ylabel("Index of Refraction n")
plt.show()
```

```python
dmin = np.array([52.15,50.32,49.22,48.47,48.24,])
print(dmin)
error = error(erd,dmin,alpha)
print(error)
n = n(dmin,alpha)
print(n)
wvlngth = np.array([404.7,435.9,546.1,577.0,599.0])
def function_to_fit(x, a, b):
    return a*x + b
plot_x = np.linspace(375.0,675.0,100)
popt, pcov = op.curve_fit(function_to_fit, wvlngth, n, sigma=error, p0=None)
print ("Fit Parameters:", popt)
print ("Fit Parameter Uncertainties:", np.sqrt(np.diag(pcov)))
plt.errorbar(wvlngth, n, error, fmt='o')
plt.plot(plot_x, function_to_fit(plot_x, *popt), 'r')
plt.title('Graph 3 Wavelength and n only Mercury\nand Curve Fit Using Scipy')
plt.xlabel("Wavelength in nm")
plt.ylabel("Index of Refraction n")
plt.show()
```
